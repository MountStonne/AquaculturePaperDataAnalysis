```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import math
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df1 = pd.read_csv("data_part1.csv", na_filter=False)
df2 = pd.read_csv("data_part2.csv", na_filter=False, header = None)
df2 = df2.drop(columns = [2])

df1 = df1.rename(columns = {'0':'text1', '1':'text2', '2':'text3', '3':'text4'})
df2 = df2.rename(columns = {0:'text1', 1:'text2'})
df = df1.append(df2)
df = df.reset_index(drop = True)
for i in range(611, 624):
    df['text3'][i] = ''
    df['text4'][i] = ''
df
```

```python
# two keywords
keywordListIndex1 = ['Aquaculture', 'Fishery', 'Fish_Hatchery', 'Maritime', 'Fish_Farm', 'Farming', 'Review']
keywordListIndex2 = ['Diving', 'Compression_Disease', 'Epidemiologic', 'Skin_Disease', 'Musculoskeletal_symptoms',\
                'Noise', 'Noise_Control', 'Hearing_Loss', 'Visual_Health', 'Eye_Health', 'Oral_Health',\
                'Respiratory_system', 'Asthma', 'Lung_Cancer', 'HIV', 'Drug', 'Opioid', 'Violence', 'Gender_Bias',\
                'Algae', 'Antibiotic', 'Cardiometabolic_Risks']
keywordList1 = ['Aquaculture', 'Fishery', 'Fish Hatchery', 'Maritime', 'Fish Farm', 'Farming', 'Review']
keywordList2 = ['Diving', 'Compression Disease', 'Epidemiologic', 'Skin Disease', 'Musculoskeletal symptoms',\
                'Noise', 'Noise Control', 'Hearing Loss', 'Visual Health', 'Eye Health', 'Oral Health',\
                'Respiratory system', 'Asthma', 'Lung Cancer', 'HIV', 'Drug', 'Opioid', 'Violence', 'Gender Bias',\
                'Algae', 'Antibiotic', 'Cardiometabolic Risks']


category = {'cate1': [], 'cate2': [], 'paper_numbers': [], 'paper_index': []}
for i in tqdm(range(len(keywordListIndex1))):
    for j in range(len(keywordListIndex2)):
        counter = 0
        indexArray = []
#         exec(keywordListIndex1[i] + keywordListIndex2[j] + "=[]")
        for k in range(len(df)):
            text1 = df.loc[k, 'text1'].upper()
            text2 = df.loc[k, 'text2'].upper()
            text3 = df.loc[k, 'text3'].upper()
            text4 = df.loc[k, 'text4'].upper()
            
            if (keywordList1[i].upper() in text1 and keywordList2[j].upper() in text1) or \
            (keywordList1[i].upper() in text2 and keywordList2[j].upper() in text2) or \
            (keywordList1[i].upper() in text3 and keywordList2[j].upper() in text3) or \
            (keywordList1[i].upper() in text4 and keywordList2[j].upper() in text3):
                counter += 1
                indexArray.append(k)
            
        category['cate1'].append(keywordList1[i])
        category['cate2'].append(keywordList2[j])
        category['paper_numbers'].append(counter)
        category['paper_index'].append(indexArray)
#                 exec(keywordListIndex1[i] + keywordListIndex2[j] + ".append(k)")
twoKeywords = pd.DataFrame.from_dict(category)
```
